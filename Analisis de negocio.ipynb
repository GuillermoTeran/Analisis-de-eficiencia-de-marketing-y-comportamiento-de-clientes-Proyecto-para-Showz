### ¬øCu√°ntas personas lo usan cada d√≠a, semana y mes?

```python
print(visits.columns)
```

```python
visits['day_visits'] = visits['start_ts'].dt.date # Agregar una columna 'dia'
daily_visits = visits.groupby('day_visits').agg({'uid': 'nunique'}).reset_index() # Calcular las visitas diarias por usuario 
daily_visits.rename(columns={'uid': 'unique_users'}, inplace=True)

visits['week_visits'] = visits['start_ts'].dt.isocalendar().week # Agregar una columna 'semana'
weekly_visits = visits.groupby('week_visits').agg({'uid': 'nunique'}).reset_index() # Calcular las visitas semanales por usuario
weekly_visits.rename(columns={'uid': 'unique_users'}, inplace=True)

visits['month_visits'] = visits['start_ts'].dt.to_period('M').dt.start_time # Agregar una columna 'mes'
monthly_visits = visits.groupby('month_visits').agg({'uid': 'nunique'}).reset_index() # Calcular las visitas mensuales por usuario
monthly_visits.rename(columns={'uid': 'unique_users'}, inplace=True)

# Graficar los resultados
plt.figure(figsize=(12, 6))

# Graficar visitas diarias
plt.subplot(3, 1, 1)
plt.plot(daily_visits['day_visits'], daily_visits['unique_users'], color='blue')
plt.title('Visitas Diarias')
plt.xlabel('Fecha')
plt.ylabel('Usuarios √önicos')

# Graficar visitas semanales
plt.subplot(3, 1, 2)
plt.plot(weekly_visits['week_visits'], weekly_visits['unique_users'], color='green')
plt.title('Visitas Semanales')
plt.xlabel('Semana')
plt.ylabel('Usuarios √önicos')

# Graficar visitas mensuales
plt.subplot(3, 1, 3)
plt.plot(monthly_visits['month_visits'], monthly_visits['unique_users'], color='red')
plt.title('Visitas Mensuales')
plt.xlabel('Mes')
plt.ylabel('Usuarios √önicos')

# Ajustar los gr√°ficos
plt.tight_layout()
plt.show()
```

```python
# Calcular visitas promedio por d√≠a, semana y mes
daily_visits_avg = daily_visits['unique_users'].mean()
weekly_visits_avg = weekly_visits['unique_users'].mean()
monthly_visits_avg = monthly_visits['unique_users'].mean()

# Crear gr√°ficos
plt.figure(figsize=(14, 10))

# Gr√°fico de visitas diarias
plt.subplot(3, 1, 1)
plt.plot(daily_visits['day_visits'], daily_visits['unique_users'], label='Visitas Diarias', color='blue')
plt.axhline(y=daily_visits_avg, color='red', linestyle='--', label=f'Promedio: {daily_visits_avg:.2f}')
plt.title('Visitas Diarias')
plt.xlabel('Fecha')
plt.ylabel('Usuarios √önicos')
plt.legend()

# Gr√°fico de visitas semanales
plt.subplot(3, 1, 2)
plt.plot(weekly_visits['week_visits'], weekly_visits['unique_users'], label='Visitas Semanales', color='green')
plt.axhline(y=weekly_visits_avg, color='red', linestyle='--', label=f'Promedio: {weekly_visits_avg:.2f}')
plt.title('Visitas Semanales')
plt.xlabel('Semana')
plt.ylabel('Usuarios √önicos')
plt.legend()

# Gr√°fico de visitas mensuales
plt.subplot(3, 1, 3)
plt.plot(monthly_visits['month_visits'], monthly_visits['unique_users'], label='Visitas Mensuales', color='purple')
plt.axhline(y=monthly_visits_avg, color='red', linestyle='--', label=f'Promedio: {monthly_visits_avg:.2f}')
plt.title('Visitas Mensuales')
plt.xlabel('Mes')
plt.ylabel('Usuarios √önicos')
plt.legend()

# Ajustar dise√±o de los gr√°ficos
plt.tight_layout()
plt.show()
```

```python
# Tabla din√°mica para visitas diarias
daily_pivot = daily_visits.pivot_table(
    index='day_visits',  
    values='unique_users',  
    aggfunc='sum'  
)

# Tabla din√°mica para visitas semanales
weekly_pivot = weekly_visits.pivot_table(
    index='week_visits',  
    values='unique_users',
    aggfunc='sum'
)

# Tabla din√°mica para visitas mensuales
monthly_pivot = monthly_visits.pivot_table(
    index='month_visits',  
    values='unique_users',
    aggfunc='sum'
)

daily_pivot
```

```python
# Crear la columna 'first_day_visit'
first_visit_day_by_customers = visits.groupby('uid')['day_visits'].min()

first_visit_day_by_customers.name = 'first_day_visits'

visits = pd.merge(visits, first_visit_day_by_customers)

# Crear la columna 'first_month_visits'
visits['first_month_visits'] = visits['first_day_visits'].dt.to_period('M')

visits_grouped_by_cohorts = visits.groupby(['first_month_visits', 'month_visits']).agg({'uid': 'nunique'})

visits_grouped_by_cohorts.pivot_table(index='first_month_visits', columns='month_visits', values='uid', aggfunc='sum')




first_visit_day_by_customers.name = 'first_visits_day'

visits = visits.join(first_visit_day_by_customers, on='uid')

visits['first_visits_month'] = visits['first_visits_day'].astype('datetime64[M]')

visits['visits_month'] = visits['visits_day'].astype('datetime64[M]')

visits_grouped_by_cohorts = visits.groupby(['first_visits_month', 'visits_month']).agg({'uid': 'nunique'})

visits_grouped_by_cohorts.pivot_table(index='first_visits_month', columns='visits_month', values='uid', aggfunc='sum')
```

```python
visits_grouped_by_cohorts = visits_grouped_by_cohorts.reset_index()
visits_grouped_by_cohorts['cohort_lifetime'] = (
    visits_grouped_by_cohorts['visits_month']
    - visits_grouped_by_cohorts['first_visits_month']
)

visits_grouped_by_cohorts['cohort_lifetime'] = visits_grouped_by_cohorts[
    'cohort_lifetime'
] / np.timedelta64(1, 'M')

visits_grouped_by_cohorts['cohort_lifetime'] = (
    visits_grouped_by_cohorts['cohort_lifetime'].round().astype('int')
)

visits_grouped_by_cohorts['first_visits_month'].dt.strftime('%Y-%m')

visits_grouped_by_cohorts_pivot = visits_grouped_by_cohorts.pivot_table(
    index='first_visits_month',
    columns='cohort_lifetime',
    values='uid',
    aggfunc='sum',
)

visits_grouped_by_cohorts_pivot
```

```python
plt.figure(figsize=(13, 9))
plt.title('Mapa de Calor de Visitas Mensuales por Duraci√≥n del Ciclo de Vida de los Usuarios', fontsize=16, pad=20)

# Crear el mapa de calor y almacenar el eje en una variable
ax = sns.heatmap(
    visits_grouped_by_cohorts_pivot,
    annot=True,
    fmt='.1f',
    linewidths=1,
    linecolor='gray',
    cmap='Blues',
    cbar_kws={'label': 'N√∫mero de Usuarios √önicos'}
)

# Formatear las etiquetas del eje Y para mostrar solo 'AAAA-MM-DD'
ax.set_yticklabels(ax.get_yticklabels(), rotation=0, fontsize=10)
ax.set_yticklabels([label.get_text().split('T')[0] for label in ax.get_yticklabels()])

# Mostrar el gr√°fico
plt.show()
```

```python
# Unimos los datasets visits y orders par obtener la actividad de los usuarios
visits_activity = orders.merge(visits, on= 'uid')
```

```python
# Conversi√≥n de la columna 'activity_day' a formato de fecha y hora
visits_activity['activity_day'] = pd.to_datetime(visits_activity['buy_ts'])

# Encontrar la primera fecha de actividad de cada usuario
first_activity_day_by_customers = visits_activity.groupby(['uid'])['activity_day'].min()
first_activity_day_by_customers.name = 'first_activity_day'
visits_activity = visits_activity.join(first_activity_day_by_customers, on='uid')

# C√°lculo del inicio de mes para cada fecha de actividad
visits_activity['activity_month'] = visits_activity['activity_day'].dt.to_period('M').dt.to_timestamp()

# C√°lculo del inicio de mes para la primera fecha de actividad de cada usuario
visits_activity['first_activity_month'] = visits_activity['first_activity_day'].dt.to_period('M').dt.to_timestamp()

visits_activity['cohort_lifetime'] = (
    visits_activity['activity_month'] - visits_activity['first_activity_month']
)
visits_activity['cohort_lifetime'] = visits_activity[
    'cohort_lifetime'
] / np.timedelta64(1, 'M')
visits_activity['cohort_lifetime'] = visits_activity['cohort_lifetime'].astype(int)

monthly_cohorts = (
    visits_activity.groupby(['first_activity_month', 'cohort_lifetime'])
    .agg({'uid': 'nunique'})
    .reset_index()
)

initial_visits_count_per_month = monthly_cohorts[monthly_cohorts['cohort_lifetime'] == 0][
    ['first_activity_month', 'uid']
]

initial_visits_count_per_month = initial_visits_count_per_month.rename(
    columns={'uid': 'cohort_visits'}
)

monthly_cohorts = monthly_cohorts.merge(initial_visits_count_per_month, on='first_activity_month')

monthly_cohorts['retention'] = monthly_cohorts['uid'] / monthly_cohorts['cohort_visits']

retention_pivot = monthly_cohorts.pivot_table(
    index='first_activity_month',
    columns='cohort_lifetime',
    values='retention',
    aggfunc='sum',
)

retention_pivot
```

```python
plt.figure(figsize=(13, 9))
plt.title('Mapa de Calor de Retenci√≥n de Visitas Mensuales por Duraci√≥n del Ciclo de Vida de los Usuarios', fontsize=16, pad=20)

# Crear el mapa de calor y almacenar el eje en una variable
ax = sns.heatmap(
    retention_pivot,
    annot=True,
    fmt='.1%',
    linewidths=1,
    linecolor='gray',
    cmap='Blues',
)

# Formatear las etiquetas del eje Y para mostrar solo 'AAAA-MM-DD'
ax.set_yticklabels(ax.get_yticklabels(), rotation=0, fontsize=10)
ax.set_yticklabels([label.get_text().split('T')[0] for label in ax.get_yticklabels()])

# Mostrar el gr√°fico
plt.show()
```

```python
monthly_cohorts['churn_rate'] = monthly_cohorts.groupby(['first_activity_month'])['uid'].pct_change()

churn_pivot = monthly_cohorts.pivot_table(
    index='first_activity_month',
    columns='cohort_lifetime',
    values='churn_rate',
    aggfunc='sum',
)

churn_pivot
```

```python
# Unimos los datasets visits y orders para obtener la actividad de los pedidos
orders_activity = orders.merge(visits, on='uid')

# Conversi√≥n de la columna 'buy_ts' a formato de fecha y hora
orders_activity['activity_order_day'] = pd.to_datetime(orders_activity['buy_ts'])

# Filtrar los datos para incluir solo los dispositivos DESKTOP y source_id == 1
orders_activity_filtered = orders_activity[(orders_activity['device'] == 'DESKTOP') & (orders_activity['source_id'] == 1)]

# Encontrar la primera fecha de actividad de cada pedido
first_activity_day_by_orders = orders_activity_filtered.groupby('uid')['activity_order_day'].min()
first_activity_day_by_orders.name = 'first_activity_order_day'
orders_activity_filtered = orders_activity_filtered.join(first_activity_day_by_orders, on='uid')

# C√°lculo del inicio de mes para cada fecha de actividad
orders_activity_filtered['activity_order_month'] = orders_activity_filtered['activity_order_day'].dt.to_period('M').dt.to_timestamp()

# C√°lculo del inicio de mes para la primera fecha de actividad de cada usuario
orders_activity_filtered['first_activity_order_month'] = orders_activity_filtered['first_activity_order_day'].dt.to_period('M').dt.to_timestamp()

# C√°lculo de la duraci√≥n del ciclo de vida del usuario (en d√≠as)
orders_activity_filtered['cohort_lifetime'] = (orders_activity_filtered['activity_order_day'] - orders_activity_filtered['first_activity_order_day']).dt.days

# Crear rangos para cohort_lifetime (0d a +7d)
orders_activity_filtered['cohort_lifetime_group'] = pd.cut(
    orders_activity_filtered['cohort_lifetime'],
    bins=[-1, 0, 1, 2, 3, 4, 5, 6, 7, np.inf],
    labels=['0d', '1d', '2d', '3d', '4d', '5d', '6d', '7d', '+7d']
)

# Agrupar por mes de la primera actividad y cohort_lifetime_group, calculando el n√∫mero de usuarios √∫nicos (uid), 
# el total de revenue y la conversi√≥n
daily_cohorts = (
    orders_activity_filtered.groupby(['first_activity_order_month', 'cohort_lifetime_group'])
    .agg({
        'uid': 'nunique',  # N√∫mero de usuarios √∫nicos
        'revenue': 'sum'  # Total de revenue por cohorte
    })
    .reset_index()
)

# Contar el n√∫mero de pedidos iniciales por mes (cohort_lifetime_group == '0d')
initial_orders_count_per_month = daily_cohorts[daily_cohorts['cohort_lifetime_group'] == '0d'][['first_activity_order_month', 'uid']]

# Renombrar la columna 'uid' a 'cohort_orders'
initial_orders_count_per_month = initial_orders_count_per_month.rename(columns={'uid': 'cohort_orders'})

# Unir los datos de pedidos iniciales con el dataset daily_cohorts
daily_cohorts = daily_cohorts.merge(initial_orders_count_per_month, on='first_activity_order_month')

# Calcular el porcentaje de conversi√≥n para cada cohorte (en formato decimal)
daily_cohorts['conversion'] = daily_cohorts['uid'] / daily_cohorts['cohort_orders']

# Crear la tabla din√°mica
conversion_pivot = daily_cohorts.pivot_table(
    index='first_activity_order_month',
    columns='cohort_lifetime_group',
    values=['conversion', 'revenue'],  # Incluir la conversi√≥n y el total de ingresos
    aggfunc={'conversion': 'sum', 'revenue': 'sum'}  # Promedio de conversi√≥n y suma de revenue
)

# Llenar valores nulos con 0 y redondear a dos decimales
conversion_pivot = conversion_pivot.fillna(0).round(2)

# Mostrar la tabla
conversion_pivot
```
